```python
import itertools
import random
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import pylab
```

```python
x = 24
y = 6
cnt = 5
```

```python
def agm(x,y,cnt):
    if(cnt==0):
        return x
    a = (x+y)/2
    g = (x*y)**0.5
    cnt-=1
    #print(cnt)
    return agm(a,g,cnt)

```

```python
res=agm(x,y,5)
res
```

```python
res=agm(6,22,cnt)
res
```

```python
#l1 = random.sample(range(1, 30), 20)
#l2 = random.sample(range(1, 30), 20)
l1 =[6, 24, 9, 15, 21, 10, 22, 17, 11, 29, 8, 26, 28, 5, 2, 18, 12, 4, 7, 20,16]
l2 =[22, 26, 5, 2, 12, 14, 10, 3, 20, 23, 9, 24, 1, 15, 17, 25, 16, 11, 7, 8,7]
res = []
for i in range(0,21):
    res.append(agm(l1[i],l2[i],cnt))
#res
l1  = np.array(l1)
l2 = np.array(l2)
res = np.array(res)
#res = np.vstack((res,np.zeros(21)))
```

```python
plt.scatter(l1,res)
plt.xlabel("Number")
plt.ylabel("Result")
plt.title('l1 vs res')
plt.show()
```

```python
plt.scatter(l2,res)
plt.xlabel("Number")
plt.ylabel("Result")
plt.title('l2 vs res')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(l1, res, l2)
ax.set_title('surface')
```
